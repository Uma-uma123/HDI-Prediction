```python
new_df = df[
    [
        "country",
        "hdi",
        "life_expectancy",
        "expec_yr_school",
        "mean_yr_school",
        "gross_inc_percap"
    ]
]
```

```python
new_df = new_df.dropna()
```

```python
new_df.dtypes
```

```python
def hdi_category(hdi):
    if hdi >= 0.800:
        return "Very High"
    elif hdi >= 0.700:
        return "High"
    elif hdi >= 0.550:
        return "Medium"
    else:
        return "Low"

new_df["HDI_Category"] = new_df["hdi"].apply(hdi_category)
```
